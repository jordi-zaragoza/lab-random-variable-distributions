Refer to the `files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv` dataset.

### Instructions

1. Get the numerical variables from our dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
marketing = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
numerical = marketing.select_dtypes(np.number)
numerical
```

2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.


```python
for col in numerical.columns:
    sns.displot(numerical[col])
    plt.show()
```

```python
# They fit the exponential distribution: TCA, Num_policies, Open_comp, MPA, CLV,
```

3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
# the log-transform does it
```

```python
numerical_logs = numerical.copy()
numerical_logs = numerical_logs[['Total Claim Amount', 'Monthly Premium Auto', 'Customer Lifetime Value']]
numerical_logs = np.log(numerical_logs)
```

```python
for col in numerical_logs.columns:
    sns.displot(numerical_logs[col])
    plt.show()
```

```python
# Total Claim Amount works fine with log
```
